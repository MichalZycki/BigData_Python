 ### 4.1 Przypomnienie, co to jest biblioteka?

Biblioteka to zbiór kodu (modułów), który usprawnia codzienne zadania. Na przykład za pomocą Seaborn możesz generować wizualizacje za pomocą tylko jednego wiersza kodu zamiast pisania całej składni od podstaw.

Biblioteki są niezależnymi zbiorami które należy pobrać i zainstalować bezpośrednio w miejscu którym pracujemy.

 ### 4.2 Jak korzystać z bibliotek?
 Aby móc wykorzystać jakąkolwiek bilbiotekę w notebooku lub IDE (Studio Code, PyCharm) należy ją uprzednio zaimportować w całości, lub konkretną klasę modułu.
 
 Dlaczego po instalacji musimy cokolwiek zaimportować? Wszystko służy optymalizacji zasobów obliczeniowych maszyn. Wszystkie biblioteki w "użyciu" w jakiś sposób przechowywane są w aktywnej pamięci komputera jako zbiór kodu do wykorzystania. Hipotetycznie gdybyśmy zainstalowali 100 bibliotek nasza maszyna była by dodatkowo obciążona tymi wszystkimi bibliotekami co jest nieefektywne. Aby zoptymalizować proces importujemy do bieżącego rozwiązania tylko te które są nam potrzebne. Np. tylko 3 ze 100.
 
 Przykład klas biblioteki "datetime"
 
 
 
 W przypadku jeśli jakaś biblioteka nie jest zainstalowana a chcemy ją zaimportować otrzymamy odpowiedni komunikat który informuje o braku modułu (biblioteka to zbiór modułów) 
 
 Przykład importu biblioteki "dask". Komunikat sugeruje że powinniśmy ją doinstalować.
 
 
 Aby wykorzystać jakąkolwiek funkcjonalość z biblioteki należy tą że biblioteke zmaterializować w kodzie. Dlatego warto zwrócić uwagę że podczas importowania dobrze jest aliasować obiekty które importujemy. Dlaczego? Ponieważ:

 ### 5.1 Co to jest dokładnie pandas?
 
 Pandas w dużym skrócie to biblioteka powstała w oparciu o "numpy" i jest bardzo rozbudowanym pakietem służącym do obróbki, analizy i manipulacji danymi w postaci dwu wymiarowej - tabelka.  ### 5.2 Rodzaje obiektów/struktur?
 
 W pandas wyróżniamy generalnie 2 rodzaje obiektów którymi są:
 - Serie (Series)
 - Ramki (Dataframes)
 
 ### Series
 Seria to najprostrzy obiekt przechowujący dane, jednowymiarowy. W dużym uproszczeniu możemy serie traktować jako kolumnę w excelu która przechowuje dane
 
 ### Dataframe
 Dataframe natomiast jest obiektem złożonym z serii. Sam dataframe możemy porównać do zestawienia lub tabeli w excelu w której występuje kilka kolumn. Dataframe możemy stworzyć za pomocą połączenia ze sobą list w pandas lub zaimportować dane do dataframeu. 
 

````python
seria = ["Jennifer","MarY","Ahmed","Susan",""] # lista imion 
print(seria)
print(type(seria))
````

````python
print(pd.Series(seria)) # zamiana listy na serie
print(type(pd.Series(seria)))
````

````python
a = ["Jennifer","MarY","Ahmed","Susan",""] # lista skladajaca sie z 5 elementow
b = ("indigo","grey","grey","gold","black") # tuple skladajacy sie z 5 elementow
c = ("indigo","grey","grey","gold") # tuple składający się z 4 elementów
# złączenie w słownik listy a i tupla b (wszystkie dane bez deklaracji zmiennej)
d = {"kolumna1":["Jennifer","MarY","Ahmed","Susan",""],"kolumna2":("indigo","grey","grey","gold","black")}
# złączenie w słownik listy a i tupla b (sama deklaracja zmiennych)
e = {"kolumna1":a,"kolumna2":b}
# złączenie w słownik listy a i tupla c (sama deklaracja zmiennych)
f = {"kolumna1":a,"kolumna2":c}
g = ["a","b","c","d","e"]
````

### CSV & TXT

```
pandas.read_csv(
                filepath_or_buffer,          
                sep=_NoDefault.no_default,   
                delimiter=None,               
                header='infer',               
                names=_NoDefault.no_default,
                index_col=None,
                usecols=None,
                dtype=None,
                skiprows=None,
                skipfooter=0,
                nrows=
                )
```                
- <mark>filepath_or_buffer</mark> - nazwa pliku, lokalizacja, lokalizacja razem z nazwą pliku
- <mark>sep=</mark> - separator kolumn
- <mark>delimiter=</mark> - alias dla separatora
- <mark>header=</mark> - informacja czy importujemy nagłówki
- <mark>names=</mark> - lista nazw kolumn
- <mark>index_col=</mark> - oznaczamy kolumne lub dodajemy serie aby utworzyć indeks
- <mark>usecols=</mark> - deklarujemy tutaj tylko subset danych do odczytu
- <mark>dtype=</mark> - możliwość zadeklarowania schematu w trakcie odczytu danych np.{‘a’: np.float64, ‘b’: np.int32, ‘c’: ‘Int64’}
- <mark>skiprows=</mark> - oznaczamy ile wierszy od górych chemy pominać
- <mark>skipfooter=</mark> - oznaczamy ile wierszy od dołu chemy pominać
- <mark>nrows=</mark> - oznaczamy ile wierszy chcemy zaimportować 
- <mark>encoding=</mark> - format kodowania plików, istotne głównie dla .txt i przeważnie "utf-8"

### Funkcje/metody rewizyjne - Df

- <mark>Df</mark> - wyświetla skrócony dataframe
- <mark>Df.head() </mark> - wyswietla fragment dataframe od początku danych, wpisując wartość w nawiasie np. Df.head(5) wskazujemy żeby program wyświetlił 5 pierwszych wierszy.
- <mark>Df.tail()</mark> - wyswietla fragment dataframe od początku danych, wpisując wartość w nawiasie np. Df.tail(7) wskazujemy żeby program wyświetlił 7 ostatnich wierszy.

- <mark>Df.size</mark> - wyświetla wielkość dataframeu, iloczyn wierszy i kolumn 
- <mark>Df.shape</mark> - wyświetla ilość wierszy oraz ilość kolumn. 
- <mark>Df.index()</mark> - wyswietla indeks
- <mark>Df.columns()</mark> - wyświetla listę kolumn w dataframe
- <mark>Df.min()</mark> - wyświetla minimalne wartości dla dataframe
- <mark>Df.max()</mark> - wyświetla maksymalne wartości dla dataframe
- <mark>Df.nlargest(n,"kolumna1")</mark> - wyświetla liste n największych wartości dla konkretnej kolumny
- <mark>Df.nsmalest(n,"kolumna1")</mark> - wyświetla liste n najmniejszych wartości dla konkretnej kolumny
- <mark>Df.sample()</mark> - wyświetla próbkę
- <mark>Df.values</mark> - wyświetla wartości

### 6.5.6 Laczenie DataFrameow

### Merge - polaczenia wg nazw kolumn oraz indeksow
```
df.merge(
         right,          
         how,   
         on,
         left_on,
         right_on,
         left_index,
         right_index,
         suffixes,
         indicator
        )
```                
- <mark>right</mark> - DataFrame ktory chcemy dodac
- <mark>how</mark> - sposob polaczenia jak w SQL {‘left’, ‘right’, ‘outer’, ‘inner’, ‘cross’}
- <mark>on</mark> - kolumna albo indeks na podstawie ktorego chcemy laczyc DataFramey, kolumna lub indeks musi wystepowac w tych samych DataFrameach 
- <mark>left_on</mark> - kolumna/kolumny wg ktorych laczymy po lewej stronie
- <mark>right_on</mark> - kolumna/kolumny wg ktorych laczymy po prawej stronie
- <mark>left_index</mark> - indeks/indeksy wg ktorych laczymy po lewej stronie
- <mark>right_index</mark> - indeks/indeksy wg ktorych laczymy po prawej stronie 
- <mark>suffixes</mark> - suffix ktory oznacza nazwy kolumn jesli sa takie same w obydwu DataFrameach dla rozroznienia np. lewy zestaw ma suffix "_x" a prawy "_y"
- <mark>indicator</mark> - kolumna do zmiany 


Dokumentacja:

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html

### Join -  polaczenia wg indeksow

df.join(

         on,
         lsuffix,
         rsuffix
        )

- <mark>other</mark> - DataFrame ktory chcemy dodac

- <mark>on</mark> - kolumna albo indeks na podstawie ktorego chcemy laczyc DataFramey, kolumna lub indeks musi wystepowac w tych samych DataFrameach 
- <mark>lsuffix</mark> - lewy suffix ktory oznacza nazwy kolumn jesli sa takie same w obydwu DataFrameach dla rozroznienia
- <mark>rsuffix</mark> - prawy suffix ktory oznacza nazwy kolumn jesli sa takie same w obydwu DataFrameach dla rozroznienia
